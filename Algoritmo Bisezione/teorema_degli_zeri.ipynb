# Algoritmo di bisezione per il calcolo degli zeri di una funzione non lineare

```python
import matplotlib.pyplot as plt
# importa a biblioteca matplotlib
# d'ora in poi si potrà usare plt invece di scrivere matplotlib.pyplot
import math
import numpy as np

f = lambda x: x**3+4*x**2-10

def alg(a,b,func):
    # TO-DO
    if (f(a) * f(b)) < 0:
        print("posso procedere a trovare quale f(x) = 0")
        # criterio di arresto
        a0, b0 = a, b
        ak, bk = 0, 0
        # punto mediano
        ck = (a0 + b0)/2
        # k = passi fatti per trovare lo zero
        k = 0
        while (f(ck) != 0):
            #print("Valore f(ck)", f(ck))
            '''
            se è la prima volta nel while il c sarà
            quello calcolato prima del while,
            se no si calcola il nuovo ck in base
            ai nuovi intervalli [ak, bk]
            '''
            
            if (k == 0):
                ak, bk = a0, b0
            else:
                ck = (ak + bk)/2
            
            if(f(ck) * f(ak) < 0):
                bk = ck
            elif(f(ck) * f(bk) < 0) :
                ak = ck
            else:
                print("TROVATO f(x) = 0 è: ", ck)
                return ck
            k += 1
            
    else:
        print("non è una funzione continua")
    
a1, b1 = -4, 2
# N numero di elementi tra a1 e b1
N = 10
c = alg(a1, b1, f)

x = np.linspace(a1, b1, N)
y = f(x)
#print(x, y)

# Creazione grafico
plt.plot(x, y, 'g-o')
plt.grid(True)
plt.xlabel('x')
# xlabel assegna un'etichetta all'asse delle ascisse
plt.ylabel('f(x)')
plt.annotate('Punto in cui f(x) = 0',xy=(c,f(c)),xytext=(-4, 4),arrowprops={'arrowstyle':'->'}) #Annotazione
plt.savefig("filename.jpg")  #Salva la figura in un file con formato grahico jpg
plt.show()
```

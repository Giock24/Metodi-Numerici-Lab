# matplotlib - Disegnare  in Python

## Introduzione

Matplotlib è un'eccellente libreria grafica 2D e 3D per la generazione di figure scientifiche. Alcuni dei numerosi vantaggi di questa libreria includono:

* Facile da inizializzare
* Grande controllo di ogni elemento in una figura.
* Output di alta qualità in molti formati, inclusi PNG, PDF, SVG, EPS e PGF.

Maggiori informazioni alla pagina web di Matplotlib: http://matplotlib.org/

Per cominciare ad usare matplotlib

```python
import matplotlib.pyplot as plt
# importa a biblioteca matplotlib
# d'ora in poi si potrà usare plt invece di scrivere matplotlib.pyplot
import math
```

### Esempio

```python
x = list(range(1,10))
y = [ i ** 2  for i in x]
# crea una lista y che contiene i quadrati degli elementi di x
y3=[i+i**2+i**3 for i in x]
# crea una lista y3 che contiene i+ i^2 + i^3 cioè una funzione cubica

print(x)
print(y)
print(y3)
```

```python
# disegna il grafico
plt.plot(x, y, 'g-o')
# plot disegna il grafico, 'ro' indica che i punti sono rossi
# la prima lettera indica il colore, la seconda il tipo di punto
# se uno mette tra le lettere un : i punti sono connessi da una linea
# se uno mette un -- i punti sono connessi da una linea tratteggiata
# se uno mette un -. i punti sono connessi da una linea mista
# se uno mette un - i punti sono connessi da una linea continua
plt.grid(True)
# grid visualizza la griglia
# di Default grid(False)
plt.xlabel('x')
# xlabel assegna un'etichetta all'asse delle ascisse
plt.ylabel('y')
plt.title('Questo è un grafico di y=x^2 e y=x+x^2+x^3')
# title assegna un titolo al grafico
plt.legend(["Funzione1","Funzione2"])
# legend visualizza la legenda
plt.savefig("filename.jpg")  #Salva la figura in un file con formato grahico jpg
plt.show()
# show visualizza il grafico

```

```python
plt.plot(x, y3, 'g*-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend(["Funzione2"])
plt.show()
```

## Disegnare due funzioni sullo stesso grafico

```python
plt.plot(x, y3, 'r:o')
plt.plot(x,y,'g*--')
plt.xlabel('x')
plt.ylabel('y')
plt.text(1,400,'Testo Libero',c='r')  #Scrivere testo libero a partire da una coordinata specificata
plt.annotate('Importante',xy=(6,250),xytext=(5.0, 400),arrowprops={'arrowstyle':'->'}) #Annotazione
plt.title('title')
plt.legend(["Funzione1","Funzione2"])
plt.show()
```

```python
x1=list(range(1,101))
x2=list(range(1,101))
y1 = [ i **(-5.0/3.0)  for i in x1]
y2 = 1 / (1 + math.e**(-x1))
# attualmente viene mostrato un grafico in cui non capisci se varia o meno in una certa zona
# sembra che sia costante
plt.plot(x1,y1,'r')
plt.show()
```

## Disegnare in scala semilogaritmica in y

```python
plt.semilogy(x1,y1,'r')  #Scala logaritmica in y
# con la scala logaritmica si riesce a vedere meglio la variazione
plt.show()
```

## Suddividere la figura in sottofigure

```python
plt.subplot(2,2,1)
# subplot divide la finestra in più parti, 1 riga e 2 colonne
# il terzo parametro indica dove disegnare il grafico
plt.plot(x, y, 'r--',linewidth=5.0)
# nella funzione plot è possibile assenare la larghezza della linea
plt.title("Funzione 1")
plt.subplot(2,2,2)
plt.plot( x,y3, 'g^-',linewidth=1.0);
plt.title("Funzione 2")
plt.subplot(2,2,3)
plt.plot(x, y, 'r--',linewidth=5.0)
plt.title("Funzione 3")
plt.subplot(2,2,4)
plt.plot(x, y, 'r--',linewidth=5.0)
plt.title("Funzione 4")
plt.savefig("filename2.png")
plt.show()
```

## Grafici a barre

```python
names = ['a', 'b', 'c', 'd']
values = [1, 10, 40, 50]
plt.bar(names, values)
plt.show()
```
